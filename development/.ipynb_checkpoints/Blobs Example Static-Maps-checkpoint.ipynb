```python
# Import required libraries
import blobs
import numpy as np
import pandas as pd
import pysal as ps
import random as rdm
from pysal.contrib.viz import mapping as maps
%matplotlib inline
from pylab import *
```

```python
shp_link = ('tracts/Master311.shp')
print 'Reading from ', shp_link

graffiti_p = np.array(ps.open(shp_link.replace('.shp', '.dbf')).by_col('graffiti_p'))
```

```python
# Before any further analysis, it is always good practice to visualize the distribution of values.
# Let's see what the distribution of 311 Calls on Graffiti looks like in Chicago.
_ = maps.plot_choropleth(shp_link, graffiti_p, 'quantiles', cmap='Greens', figsize=(7, 9))
_ = maps.plot_choropleth(shp_link, graffiti_p, 'equal_interval', cmap='Greens', figsize=(7, 9))
```

```python
# Next, let's run a LISA (local indicator of spatial autocorrelation) on Graffiti 311 calls.
# First, we need to create a spatial weight file. We'll use a simple contiguity matrix, and 
# then run the LISA.
w = ps.queen_from_shapefile(shp_link)
lisa = ps.Moran_Local(graffiti_p, w, permutations=9999)
```

```python
# Okay, let's look at the LISA we created.
_ = maps.plot_lisa_cluster(shp_link, lisa)
```

```python
shp_link = ('tracts/Master311.shp')
dbf = ps.open('tracts/Master311.dbf')
cols = np.array([dbf.by_col(col) for col in dbf.header]).T
df = pd.DataFrame(cols)
df.columns = dbf.header
df.columns = df.columns.map(lambda x: x.lower())
df.commarea = df.commarea.astype('int')
df['order'] = df.index
cols = np.array([dbf.by_col(col) for col in dbf.header]).T
values = np.array(ps.open(('tracts/Master311.dbf')).by_col('graffiti_p'))

types = ['quantiles', 'equal_interval']
for typ in types:
    maps.plot_choropleth(shp_link, values, typ, title=typ, figsize=(7,9))
```

```python
# Format data and merge on shapefile IDs
ordered_tracts = pd.DataFrame(df.loc[:,['tractce10', 'commarea', 'order']])
calls = pd.merge(init_calls, ordered_tracts, how='right', left_on='tractID', 
    right_on='tractce10', sort=False).fillna(0).sort(['order'])
calls = calls.drop(['order', 'commarea'],1)

class bd:
  data = calls
  w = w
  shp_link = shp_link
  id = 'tractce10'
  level = 'tract'
```

```python
# Assign the attribute data.
d = bd()
```

## Create Blobs

You can assign what attribute you would like to "blob" accordingly. In this example, blobs are created from census tracts and must have a minimum of 10,000 persons per blob.

```python
# Create blobs minimum population of 10,000 in each blob. This calls a Pop-Up Window with the Map.
b = blobs.Blobs(d, 'pop', 10000, iterations=1)
```

```python
# Cluster the blobs along similarities in the data
cl = blobs.Cluster_blobs(b, blobs_per_cluster=10)
```

```python
# Create blobs with a minimum of 30 tracts in each blob, and cluster
b = blobs.Blobs(d, 'areas', 30)
cl = blobs.Cluster_blobs(b, blobs_per_cluster=10)
```

```python
# Have around 3 blobs per cluster
cl.set_n_clusters(3)
```

```python
# See blob assignments
cl.assignments
```

```python
# Plot blobs along one of the variables
b.plot_blobs('all_calls_per1000_mean')
```
